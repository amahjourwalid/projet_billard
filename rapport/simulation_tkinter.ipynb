```python
from tkinter import* 
import math



# initialisation of a basic window with minimize button ...

angle = 0
speed = 20
WIDTH = 600
HEIGHT = 500
X1rec = WIDTH/2 - 10
Y1rec = HEIGHT/2 - 10
X2rec = X1rec + 20
Y2rec = Y1rec + 20
X1Line = 0
Y1Line = 0
X2Line = 0
Y2Line = 0

button = False

def calculateLinePosition():
    global X1Line
    X1Line = X1rec + (X2rec - X1rec)/2
    global Y1Line
    Y1Line = Y1rec + (Y2rec - Y1rec)/2
    global X2Line
    X2Line = X1Line + 100*math.sin(angle*math.pi/180)
    global Y2Line
    Y2Line = Y1Line + 100*math.cos(angle*math.pi/180)
    
    
calculateLinePosition()
root = Tk()#creating and openning a window
root.title("Billard Simulation")
root.geometry("800x600")#size of the window 
#creatinf a frame for the buttons in th bottom

bottomFrame=Frame(root)
bottomFrame.pack(side=BOTTOM)

#giv a name to our window

#thelabel=Label(root,text="billard simulation") 
#thelabel.pack()

#creating buttons (pause and restart buttons )
def pause(event):#function of the button 
    global button
    button = False
    
Pause=Button(bottomFrame,text="Pause",fg="Blue")
Pause.bind(" <Button-1>",pause)
Pause.pack(fill=X)
def Start(event):#function of the button
    global button
    button = True
restart=Button(bottomFrame,text="Start/Restart",fg="Green")
restart.bind("<Button-1>",Start)
restart.pack(fill=X)

#creating entries ( angle)

label1=Label(root,text="please choose an angle to start with")#title of the entry
label1.pack()

#obtaining data from our entry  

def evaluateangle(event):
    print("Evaluatin angle")
    global angle
    
    
    entry = entrySpace.get()
    if entry :
        angle = float(entry)
        ans.configure(text="Your angle is " +str(angle))
        ans.pack()
        calculateLinePosition()
        canvas.delete('all')
        canvas.create_rectangle(0, 0, WIDTH, HEIGHT, fill="Green")
        canvas.create_oval(X1rec, Y1rec, X2rec, Y2rec , width = 0, fill = 'white')
        canvas.create_line(X1Line, Y1Line, X2Line, Y2Line, width= 2,fill = 'white')
        canvas.pack()
    
entrySpace=Entry(root)#creating an entry space    
entrySpace.bind("<Return>",evaluateangle)
entrySpace.pack()
ans=Label(root)
ans.configure(text="Your initial angle is " +str(angle))
ans.pack()


canvas=Canvas(root,width=WIDTH,height=HEIGHT)
canvas.pack()
canvas.create_rectangle(0,0,WIDTH,HEIGHT,fill="Green")
canvas.create_oval(X1rec, Y1rec, X2rec, Y2rec , width = 0, fill = 'white')
canvas.create_line(X1Line, Y1Line, X2Line, Y2Line, width= 2,fill = 'white')
canvas.pack()
time = 60
def refresh():
    if(button):
        global X1rec
        global X2rec
        global Y1rec
        global Y2rec
        global speed
        global angle
        global HEIGHT
        global WIDTH
        if X1rec - float(speed)*math.sin(angle*math.pi/180) >= 0 :
            if X1rec - float(speed)*math.sin(angle*math.pi/180) > WIDTH - 20:
                if angle >= 90 :
                    angle = 360 - angle
                else :
                    angle = angle + 270
            X1rec = X1rec - float(speed)*math.sin(angle*math.pi/180)
            X2rec = X1rec + 20
        else :
            if angle >= 90 :
                angle = (360 - angle)%360
            else :
                angle = (angle + 270)%360
            X1rec = X1rec - float(speed)*math.sin(angle*math.pi/180)
            X2rec = X1rec + 20
        if Y1rec - float(speed)*math.cos(angle*math.pi/180) >= 0:
            if Y1rec - float(speed)*math.cos(angle*math.pi/180) > HEIGHT - 40:
                if angle <= 180 :
                    angle = (360 + 180 - angle)%360
                else :
                    angle = (180 - angle)%360
            Y1rec = Y1rec - float(speed)*math.cos(angle*math.pi/180)
            Y2rec = Y1rec + 20
        else :
            if angle <= 90 :
                angle = (180 - angle)%360
            else :
                angle = (angle + 270)%360
            Y1rec = Y1rec - float(speed)*math.cos(angle*math.pi/180)
            Y2rec = Y1rec + 20
        ans.configure(text="Your angle is " +str(angle))
        ans.pack()
        calculateLinePosition()
        canvas.delete('all')
        canvas.create_rectangle(0, 0, WIDTH, HEIGHT, fill="Green")
        canvas.create_oval(X1rec, Y1rec, X2rec, Y2rec , width = 0, fill = 'white')
        canvas.create_line(X1Line, Y1Line, X2Line, Y2Line, width= 2,fill = 'white')
        canvas.pack()
    #canvas.after(1000,refresh)
    canvas.after(100,refresh)

canvas.after(1,refresh)
root.mainloop()#to maintain the window open 
```
